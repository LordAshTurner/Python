## If-Statements with User-Defined Functions

```python
# create smaller subset of data
df = ords_prods_merge[:1000000]
```

```python
# function for categorizing prices
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# create new column with price_label output
df['price_range'] = df.apply(price_label, axis=1)
```

## Step 4.
When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Create list orders hour_of_day by value count
ords_prods_merge['order_hour_of_day'].value_counts()
```

The values in order_hour_of_day are as follows for busiest_period_of_day:
Most Orders = 10, 11, 14, 15, 13, 12, 16 and 9.
Average Orders = 17, 8, 18, 19, 20, 7, 21 and 22.
Fewest Orders = 23, 6, 0, 1, 5, 2, 4 and 3.

```python
# Create for-loop statement for a new column called “busiest_period_of_day” 
busiest_period_of_day = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    busiest_period_of_day.append("Most Orders")
  elif value in [17, 8, 18, 19, 20, 7 ,21, 22]:
    busiest_period_of_day.append("Average Orders")
  else:
    busiest_period_of_day.append("Fewest Orders")
```

```python
# Create new column busiest_period_of_day in ords_prods_merge
ords_prods_merge['busiest_period_of_day'] = busiest_period_of_day
```
